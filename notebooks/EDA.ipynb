# Exploratory Data Analysis — US Vehicle Listings

This notebook explores the `vehicles_us.csv` dataset to understand the distribution of vehicle prices, conditions, mileage, and other attributes.

```python
import pandas as pd
import plotly.express as px
```

## 1. Load the Data

```python
df = pd.read_csv('../vehicles_us.csv')
print('Shape:', df.shape)
df.head()
```

## 2. General Overview

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

## 3. Price Distribution

```python
fig = px.histogram(
    df, x='price', nbins=50,
    title='Distribution of Vehicle Prices',
    labels={'price': 'Price (USD)'},
    range_x=[0, 100000]
)
fig.show()
```

## 4. Odometer Distribution

```python
fig = px.histogram(
    df, x='odometer', nbins=50,
    title='Distribution of Odometer Readings',
    labels={'odometer': 'Odometer (miles)'}
)
fig.show()
```

## 5. Price vs Odometer

```python
fig = px.scatter(
    df, x='odometer', y='price', color='condition',
    title='Price vs Odometer by Condition',
    labels={'odometer': 'Odometer (miles)', 'price': 'Price (USD)'},
    range_y=[0, 100000], opacity=0.4
)
fig.show()
```

## 6. Vehicle Type Distribution

```python
type_counts = df['type'].value_counts().reset_index()
type_counts.columns = ['type', 'count']

fig = px.bar(
    type_counts, x='type', y='count',
    title='Number of Listings by Vehicle Type',
    labels={'type': 'Vehicle Type', 'count': 'Number of Listings'}
)
fig.show()
```

## 7. Median Price by Vehicle Condition

```python
condition_order = ['new', 'like new', 'excellent', 'good', 'fair', 'salvage']
median_price = (
    df.groupby('condition')['price']
    .median()
    .reindex(condition_order)
    .reset_index()
)
median_price.columns = ['condition', 'median_price']

fig = px.bar(
    median_price, x='condition', y='median_price',
    title='Median Price by Vehicle Condition',
    labels={'condition': 'Condition', 'median_price': 'Median Price (USD)'}
)
fig.show()
```

## 8. Model Year Distribution

```python
fig = px.histogram(
    df, x='model_year', nbins=40,
    title='Distribution of Model Years',
    labels={'model_year': 'Model Year'}
)
fig.show()
```
